```python
from bs4 import BeautifulSoup
import re
```

```python
import pandas as pd
import numpy as np
```

```python
HospAdmissionsFollowedByHospReadmission = pd.read_csv("../data/MIMICIV_3_1_HospAdmissions_FollowedBy_HospReadmission_fromDischargeNotes.csv")
```

```python
HospDischargeNotes = pd.read_csv("../data/discharge.csv")
```

```python
Merged_DischargeNotes_ReadmissionEnsue = pd.merge(HospDischargeNotes, HospAdmissionsFollowedByHospReadmission, on=['subject_id', 'hadm_id'], how='left').sort_values(by=['subject_id', 'hadm_id'])
Merged_DischargeNotes_ReadmissionEnsue
```

```python
Merged_DischargeNotes_ReadmissionEnsue['ReadmissionEnsue'] = np.where(Merged_DischargeNotes_ReadmissionEnsue['AdminNum'].isna(), 0, 1)
Merged_DischargeNotes_ReadmissionEnsue
```

```python
Merged_DischargeNotes_ReadmissionEnsue_sparse = Merged_DischargeNotes_ReadmissionEnsue[['text', 'ReadmissionEnsue']]
Merged_DischargeNotes_ReadmissionEnsue_sparse
```

```python
Merged_DischargeNotes_ReadmissionEnsue_sparse['text'] = Merged_DischargeNotes_ReadmissionEnsue_sparse['text'].str.replace(r'\s*_{1,}\s*Unit\s*No:\s*{1,}\s*\n\n\s*Admission\s*Date:\s*_{1,}\s*Discharge\s*Date:\s*_{1,}\s*\n\n\s*Date\s*of\s*\Birth:\s*_{1,}\s*', '', regex=True)
Merged_DischargeNotes_ReadmissionEnsue_sparse
```
