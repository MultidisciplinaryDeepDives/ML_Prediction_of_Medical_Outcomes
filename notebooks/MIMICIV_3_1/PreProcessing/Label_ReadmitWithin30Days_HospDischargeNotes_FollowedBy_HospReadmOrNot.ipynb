```python
import pandas as pd
import numpy as np
```

```python
HospAdmissionsFollowedByHospReadmissionWithin30Days = pd.read_csv("../data/Find_HospReadmissions_within30days.csv")
```

```python
HospAdmissionsFollowedByHospReadmissionWithin30Days = HospAdmissionsFollowedByHospReadmissionWithin30Days.drop(['Days_Interval'], axis=1)
HospAdmissionsFollowedByHospReadmissionWithin30Days
```

```python
HospAdmissionsFollowedByHospReadmissionWithin30Days['Readmission_Within_30Days'] = HospAdmissionsFollowedByHospReadmissionWithin30Days['Readmission_Within_30Days'].shift(-1)
HospAdmissionsFollowedByHospReadmissionWithin30Days
```

```python
HospAdmissionsFollowedByHospReadmission = HospAdmissionsFollowedByHospReadmissionWithin30Days
```

```python
HospDischargeNotes = pd.read_csv("../data/discharge.csv")
```

```python
Merged_DischargeNotes_ReadmissionEnsue = pd.merge(HospDischargeNotes, HospAdmissionsFollowedByHospReadmission, on=['subject_id', 'hadm_id'], how='left').sort_values(by=['subject_id', 'hadm_id'])
Merged_DischargeNotes_ReadmissionEnsue
```

Merged_DischargeNotes_ReadmissionEnsue['ReadmissionEnsue'] = np.where(Merged_DischargeNotes_ReadmissionEnsue['Readmission_Within_30Days'].isna(), 0, 1)
Merged_DischargeNotes_ReadmissionEnsue

```python
Merged_DischargeNotes_ReadmissionEnsue.rename(columns={'Readmission_Within_30Days': 'ReadmissionEnsue'}, inplace=True)
```

```python
Merged_DischargeNotes_ReadmissionEnsue_sparse = Merged_DischargeNotes_ReadmissionEnsue[['text', 'ReadmissionEnsue']].fillna(0)
Merged_DischargeNotes_ReadmissionEnsue_sparse 
```

Merged_DischargeNotes_ReadmissionEnsue_sparse['text'] = Merged_DischargeNotes_ReadmissionEnsue_sparse['text'].str.replace(r'\s*_{1,}\s*Unit\s*No:\s*{1,}\s*\n\n\s*Admission\s*Date:\s*_{1,}\s*Discharge\s*Date:\s*_{1,}\s*\n\n\s*Date\s*of\s*\Birth:\s*_{1,}\s*', '', regex=True)
Merged_DischargeNotes_ReadmissionEnsue_sparse

```python
Merged_DischargeNotes_ReadmissionEnsue_sparse.to_csv("../data/Merged_DischargeNotes_ReadmissionEnsue_sparse.csv", index=False)
```
