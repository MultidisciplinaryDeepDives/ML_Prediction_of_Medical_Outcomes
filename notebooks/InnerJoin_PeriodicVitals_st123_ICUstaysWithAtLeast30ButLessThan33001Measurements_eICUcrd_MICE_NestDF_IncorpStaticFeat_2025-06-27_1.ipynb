### Construct Nested Data Frame by ICU Stay

```python
import pandas as pd
import numpy as np
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

def create_nested_PatientUnitStay_DataRowframe(filename = '../data/df_imputedVitalsSt123_ApacheIVaTop6MortalityFactors.csv'):
    """
    Creates a DataFrame where each row represents a patient and contains their
    complete time series as nested lists. This approach maintains the grouped
    structure while providing DataFrame functionality.
    
    Think of this like creating a patient registry where each patient has a
    complete medical chart attached to their record. The DataFrame gives us
    the patient-level structure, while the nested lists preserve the detailed
    time series data.
    
    Args:
        filename (str): Path to the ICU patient CSV file
        
    Returns:
        pandas.DataFrame: DataFrame with patients as rows and time series as nested lists
    """
    
    # First, let's read the original data to understand its structure
    print("Step 1: Reading and understanding the source data...")
    raw_df = pd.read_csv(filename)
    print(f"Original data shape: {raw_df.shape}")
    print(f"Columns: {list(raw_df.columns)}")
    
    # Ensure data is properly sorted by patient and time
    # This is crucial for maintaining chronological order within each patient's data
    raw_df = raw_df.sort_values(['patientunitstayid', 'observationoffset'])
    
    print(f"\nStep 2: Grouping patients and creating time series lists...")
    
    # Create a new DataFrame structure where each patient becomes one row
    PatientUnitStay_Summaries = []
    
    # Group by patient ID and process each patient's complete time series
    for PatientUnitStay_ID, PatientUnitStay_DataRow in raw_df.groupby('patientunitstayid'):
        
        # Convert the patient's time series data to lists
        # We'll create separate lists for each measurement type
        time_pt_inMinSinceICUadmission = PatientUnitStay_DataRow['observationoffset'].tolist()
        temperatures = PatientUnitStay_DataRow['temperature'].tolist()
        sao2_values = PatientUnitStay_DataRow['sao2'].tolist()
        heart_rates = PatientUnitStay_DataRow['heartrate'].tolist()
        respirations = PatientUnitStay_DataRow['respiration'].tolist()
        cvp_values = PatientUnitStay_DataRow['cvp'].tolist()
        sys_bp = PatientUnitStay_DataRow['systemicsystolic'].tolist()
        dia_bp = PatientUnitStay_DataRow['systemicdiastolic'].tolist()
        mean_bp = PatientUnitStay_DataRow['systemicmean'].tolist()
        pa_sys = PatientUnitStay_DataRow['pasystolic'].tolist()
        pa_dia = PatientUnitStay_DataRow['padiastolic'].tolist()
        pa_mean = PatientUnitStay_DataRow['pamean'].tolist()
        st1_values = PatientUnitStay_DataRow['st1'].tolist()
        st2_values = PatientUnitStay_DataRow['st2'].tolist()
        st3_values = PatientUnitStay_DataRow['st3'].tolist()
        gender_nested = PatientUnitStay_DataRow['gender'].tolist()
        oobintubday1_nested = PatientUnitStay_DataRow['oobintubday1'].tolist()
        activetx_nested = PatientUnitStay_DataRow['activetx'].tolist()
      #  admitdiagnosis_nested = PatientUnitStay_DataRow['admitdiagnosis'].tolist()
        oobventday1_nested = PatientUnitStay_DataRow['oobventday1'].tolist()
        motor_gcs_nested = PatientUnitStay_DataRow['motor-gcs'].tolist()
        verbal_gcs_nested = PatientUnitStay_DataRow['verbal-gcs'].tolist()
      #  apachescore_nested = PatientUnitStay_DataRow['apachescore'].tolist()

        # Calculate summary statistics for this patient
        # These help us understand the patient's overall condition and monitoring duration
        num_measurements = len(PatientUnitStay_DataRow)
        monitoring_duration_inMinutes = time_pt_inMinSinceICUadmission[-1] - time_pt_inMinSinceICUadmission[0] if len(time_pt_inMinSinceICUadmission) > 1 else 0
        avg_measurement_interval = monitoring_duration_inMinutes / (num_measurements - 1) if num_measurements > 1 else 0
        
        # Calculate clinical summary statistics
        avg_temperature = np.mean(temperatures)
        avg_heart_rate = np.mean(heart_rates)
        avg_sao2 = np.mean(sao2_values)
        avg_sys_bp = np.mean(sys_bp)
        avg_dia_bp = np.mean(dia_bp)
        temperature_range = max(temperatures) - min(temperatures)
        hr_variability = np.std(heart_rates)
        sao2_variability = np.std(sao2_values)

        # Create a comprehensive patient record
        patient_record = {
            'PatientUnitStay_ID': PatientUnitStay_ID,
            'num_measurements': num_measurements,
            'monitoring_start': time_pt_inMinSinceICUadmission[0],
            'monitoring_end': time_pt_inMinSinceICUadmission[-1],
            'monitoring_duration_inMinutes_minutes': monitoring_duration_inMinutes,
            'avg_measurement_interval': avg_measurement_interval,
            
            # Clinical summary statistics
            'avg_temperature': avg_temperature,
            'avg_heart_rate': avg_heart_rate,
            'avg_saO2': avg_sao2,
            'avg_systolic_bp': avg_sys_bp,
            'avg_diastolic_bp': avg_dia_bp,
            'temperature_range': temperature_range,
            'heart_rate_variability': hr_variability,
            'saO2_variability': sao2_variability,

            # Complete time series data as lists
            'time_series_offset': time_pt_inMinSinceICUadmission,
            'time_series_temperature': temperatures,
            'time_series_sao2': sao2_values,
            'time_series_heartrate': heart_rates,
            'time_series_respiration': respirations,
            'time_series_cvp': cvp_values,
            'time_series_systolic_bp': sys_bp,
            'time_series_diastolic_bp': dia_bp,
            'time_series_mean_bp': mean_bp,
            'time_series_pa_systolic': pa_sys,
            'time_series_pa_diastolic': pa_dia,
            'time_series_pa_mean': pa_mean,
            'time_series_st1': st1_values,
            'time_series_st2': st2_values,
            'time_series_st3': st3_values,
            'time_series_nested_gender': gender_nested,
            'time_series_oobintubday1_uponICUadmission': oobintubday1_nested,
            'time_series_activetx_uponICUadmission': activetx_nested,
           # 'time_series_admitdiagnosis': admitdiagnosis_nested,
            'time_series_oobventday1_uponICUadmission': oobventday1_nested,
            'time_series_motor_gcs_uponICUadmission': motor_gcs_nested,
            'time_series_verbal_gcs_uponICUadmission': verbal_gcs_nested #,
           # 'time_series_apachescore_uponICUadmission': apachescore_nested
        }
        
        PatientUnitStay_Summaries.append(patient_record)
    
    # Create the final DataFrame
    nested_df = pd.DataFrame(PatientUnitStay_Summaries)
    
    print(f"Step 3: Created nested DataFrame with shape: {nested_df.shape}")
    print(f"Each row represents one patient with their complete time series")
    
    return nested_df
```

```python
Nested_df_imputed_1st48hrs_st123ICUstaysPeriodicVitals = pd.DataFrame(create_nested_PatientUnitStay_DataRowframe(filename = '../data/df_imputedVitalsSt123_ApacheIVaTop6MortalityFactors.csv'))
```
