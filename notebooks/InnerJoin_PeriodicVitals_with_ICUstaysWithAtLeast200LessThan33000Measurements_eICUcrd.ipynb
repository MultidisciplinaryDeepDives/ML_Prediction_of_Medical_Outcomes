```python
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.sort_values(by=['patientunitstayid', 'observationoffset']) #.drop(columns=['vitalperiodicid'])
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements
```

```python
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['Temp_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['temperature'].ffill()
```

```python
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['saO2_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['sao2'].ffill()
```

```python
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['HR_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['heartrate'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['RR_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['respiration'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['CVP_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['cvp'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['EtCO2_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['etco2'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['SysSystolic_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['systemicsystolic'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['SysDiastolic_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['systemicdiastolic'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['SysMeanBP_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['systemicmean'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['PAsystolic_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['pasystolic'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['ST1_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['st1'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['ST2_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['st2'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['ST3_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['st3'].ffill()
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements['ICP_filled'] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby('patientunitstayid')['icp'].ffill()
```

```python
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements[['HR_ff_MeanImp', 
           'RR_ff_MeanImp', 
           'CVP_ff_MeanImp', 
           'EtCO2_ff_MeanImp', 
           'SysSystolic_ff_MeanImp', 
           'SysDiastolic_ff_MeanImp', 
           'SysMeanBP_ff_MeanImp', 
           'PAsystolic_ff_MeanImp', 
           'ST1_ff_MeanImp', 
           'ST2_ff_MeanImp', 
           'ST3_ff_MeanImp', 
           'ICP_ff_MeanImp']] = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.groupby(['patientunitstayid'])['HR_filled', 
                                                                         'RR_filled', 
                                                                         'CVP_filled', 
                                                                         'EtCO2_filled', 
                                                                         'SysSystolic_filled', 
                                                                         'SysDiastolic_filled', 
                                                                         'SysMeanBP_filled', 
                                                                         'PAsystolic_filled', 
                                                                         'ST1_filled', 
                                                                         'ST2_filled', 
                                                                         'ST3_filled', 
                                                                         'ICP_filled'].transform(lambda x: x.fillna(x.mean()))
```

```python
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements = Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.drop(columns=['temperature', 'sao2', 'heartrate', 'respiration', 'cvp', 'etco2',
                                     'systemicsystolic', 'systemicdiastolic', 'systemicmean', 'pasystolic', 'st1', 'st2', 'st3', 'icp',
                                     'HR_filled', 'RR_filled', 'CVP_filled', 'EtCO2_filled', 'SysSystolic_filled', 'SysDiastolic_filled', 'SysMeanBP_filled', 'PAsystolic_filled', 'ST1_filled', 'ST2_filled', 'ST3_filled', 'ICP_filled'])    
```

```python
Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements.to_csv("../data/Merged_ICU_PeriodicVitals_WithAtLeast200ButLessThan33000Measurements_FF_MeanImp.csv", index=False)
```
